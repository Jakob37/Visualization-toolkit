```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('../data/log2.csv', index_col=0)
df.index = df.index.to_series().astype(str)
df.head()
```

```python
rows = ['4', '5', '6']
s1_i = [3,4,5]
s2_i = [6,7,8]

s1_df = df.ix[rows, s1_i]
s2_df = df.ix[rows, s2_i]

big_df = pd.DataFrame({'numbers': [],
                       'features': [],
                       'sample': []})

for col in s1_df:
    col_data = s1_df[col]
    slice_df = pd.DataFrame({'numbers':col_data, 
                             'features':rows, 
                             'sample':"s1"})
    big_df = pd.concat([big_df, slice_df])

for col in s2_df:
    col_data = s2_df[col]
    slice_df = pd.DataFrame({'numbers':col_data, 
                             'features':rows, 
                             'sample':"s2"})
    big_df = pd.concat([big_df, slice_df])

big_df
```

```python
sns.stripplot(data=big_df, x="features", y="numbers", hue="sample")
sns.plt.show()
```

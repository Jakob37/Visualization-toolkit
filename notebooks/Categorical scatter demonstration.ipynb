```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('../data/log2.csv', index_col=0)
df.index = df.index.to_series().astype(str)
df.head()
```

```python
row = '4'
s1_i = [3,4,5]
s2_i = [6,7,8]

s1_val = df.ix[row, s1_i]
s2_val = df.ix[row, s2_i]
s1_val.name = 's_data'
s2_val.name = 's_data'

s1_df = s1_val.to_frame()
s1_df['sample'] = 's1'

s2_df = s2_val.to_frame()
s2_df['sample'] = 's2'

s_df = pd.concat([s1_df, s2_df])
s_df
```

```python
sns.stripplot(data=s_df, x="sample", y="s_data")
sns.plt.show()
```
